```python
import pybullet as p
import pybullet_data
import time
import math

p.connect(p.GUI)
p.setAdditionalSearchPath(pybullet_data.getDataPath())

p.setGravity(0, 0, -9.81)
p.loadURDF("plane.urdf")
robotId = p.loadURDF(
    "/home/sha/Downloads/fr3_description/urdf/fr3.urdf", useFixedBase=True
)

numJoints = p.getNumJoints(robotId)
print(f"Joints: {numJoints}")

home_positions = [0, 0, 0, -1.57, 0, 1.57, 0.7, 0.04, 0.04]
for i in range(min(numJoints, len(home_positions))):
    p.resetJointState(robotId, i, home_positions[i])


def animate_sine(duration):
    start_time = time.time()
    amplitudes = [0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.5, 0.02, 0.02]
    frequencies = [0.5, 0.3, 0.4, 0.2, 0.6, 0.3, 0.4, 0.2, 0.2]
    phases = [0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0, 0, 0]

    while time.time() - start_time < duration:
        elapsed = time.time() - start_time
        for i in range(min(numJoints, len(home_positions))):
            pos = home_positions[i] + amplitudes[i] * math.sin(
                frequencies[i] * elapsed + phases[i]
            )
            p.setJointMotorControl2(
                robotId,
                i,
                p.POSITION_CONTROL,
                targetPosition=pos,
                force=100.0,
                maxVelocity=2.0,
            )
        p.stepSimulation()
        time.sleep(1.0 / 240.0)


def move_to(target_pos, duration=2.0):
    start_time = time.time()
    current_pos = [
        p.getJointState(robotId, i)[0] for i in range(min(numJoints, len(target_pos)))
    ]

    while time.time() - start_time < duration:
        progress = min((time.time() - start_time) / duration, 1.0)
        for i in range(min(numJoints, len(target_pos))):
            interp = current_pos[i] + progress * (target_pos[i] - current_pos[i])
            p.setJointMotorControl2(
                robotId,
                i,
                p.POSITION_CONTROL,
                targetPosition=interp,
                force=100.0,
                maxVelocity=1.0,
            )
        p.stepSimulation()
        time.sleep(1.0 / 240.0)


animate_sine(5.0)

poses = [
    [0, -0.3, 0, -1.8, 0, 1.5, 0.7, 0.01, 0.01],
    [0, 0.5, 0, -0.8, 0, 1.0, 0.7, 0.04, 0.04],
    [1.0, 0, 0, -1.5, 0, 1.5, 0.0, 0.04, 0.04],
]

for pose in poses:
    move_to(pose)
    time.sleep(1.0)

move_to(home_positions)

for _ in range(3):
    for width in [0.04, 0.01]:
        p.setJointMotorControl2(robotId, 7, p.POSITION_CONTROL, width, force=10)
        p.setJointMotorControl2(robotId, 8, p.POSITION_CONTROL, width, force=10)
        for _ in range(50):
            p.stepSimulation()
            time.sleep(1.0 / 240.0)

p.disconnect()
```
